```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import pandas as pd

# Sample DataFrame for illustration (Replace this with your actual data)
data = {
    'City': ['Jakarta', 'Yogyakarta', 'Palembang', 'Denpasar', 'Medan'],
    'Lat': [-6.1751, -7.7972, -2.9909, -8.4095, -3.5952],
    'Long': [106.8650, 110.3718, 104.7833, 115.1889, 98.6722]
}

df = pd.DataFrame(data)

# Create a figure and a map focused on Indonesia
plt.figure(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=-11, urcrnrlat=6, llcrnrlon=95, urcrnrlon=141, resolution='i')

# Draw coastlines and countries
m.drawcoastlines()
m.drawcountries()

# Convert latitude and longitude to the map projection coordinates
x, y = m(df['Long'].values, df['Lat'].values)

# Plot the points on the map
m.scatter(x, y, marker='o', color='red', zorder=10)

# Annotate the points with city names
for i, place in enumerate(df['City']):
    plt.text(x[i], y[i], place, fontsize=12, ha='right', va='bottom', zorder=15)

# Display the plot
plt.show()
```
